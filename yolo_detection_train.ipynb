## k-fold 교차 검증 모델 성능 확인

```python
import glob
from ultralytics import YOLO

model_dict = {
    'yolov8n': 'yolov8n.pt',
    'yolo11s': 'yolo11s.pt'
}

ksplit = 5
project = '/content/drive/MyDrive/ingredient project/' # 결과가 저장될 디렉토리 경로
batch = 32
epochs = 150

for name, model_v in model_dict.items():
    for k in range(ksplit):
        dataset_yaml = f'/content/drive/MyDrive/ingredient project/5_Fold/split_{k+1}/split_{k+1}_dataset.yaml'
        model = YOLO(model_v) # 모델을 초기 상태로 다시 불러옴(가중치 초기화)
        model.train(data=dataset_yaml, epochs=epochs, batch=batch, project=project, imgsz=320, name = f'{name}_fold{k+1}_train')

        # # 학습 완료후 가장 성능이 좋은 모델의 가중치 불러오기
        # best_model = YOLO(f'{project}{name}_fold{k+1}_train/weights/best.pt')

        # test_files = glob.glob(f'/content/drive/MyDrive/ingredient project/5_Fold/split_{k+1}/test/images/*.*')
        # best_model.predict(test_files,
        #                    save=True,
        #                    imgsz=320,
        #                    conf=0.3, # conf: 객체 탐지 정확성 임계값
        #                    project = '/content/drive/MyDrive/ingredient project/',
        #                    name=f'{name}_fold{k+1}_predict')
```

- 모델 성능 비교

```python
from ultralytics import YOLO

yolo11s_result_list = []
yolo8n_result_list = []

for i in ['v8n', '11s']:
    for n in range(5):
        model = YOLO(f'/content/drive/MyDrive/ingredient project/yolo{i}_fold{n+1}_train/weights/best.pt')
        result = {
            'fold': f'yolo{i}_fold{n+1}',
            'precision': round(model.val().box.mp, 3),
            'recall': round(model.val().box.mr, 3),
            'mAP50': round(model.val().box.map50, 3),
            'mAP50-95': round(model.val().box.map, 3)
        }
        if i == 'v8n':
            yolo8n_result_list.append(result)
        else:
            yolo11s_result_list.append(result)
```

```python
import pandas as pd

yolo8n_df = pd.DataFrame(yolo8n_result_list)
yolo8n_mean_df = pd.DataFrame([round(yolo8n_df[['precision', 'recall', 'mAP50', 'mAP50-95']].mean(), 3)])
yolo8n_mean_df.index = ['yolov8n_kfold_mean']
yolo8n_df.set_index('fold', inplace=True) # fold 컬럼을 index로 설정
pd.concat([yolo8n_df, yolo8n_mean_df])
```

```python
yolo11s_df = pd.DataFrame(yolo11s_result_list)
yolo11s_mean_df = pd.DataFrame([round(yolo11s_df[['precision', 'recall', 'mAP50', 'mAP50-95']].mean(), 3)])
yolo11s_mean_df.index = ['yolo11s_kfold_mean']
yolo11s_df.set_index('fold', inplace=True)
pd.concat([yolo11s_df, yolo11s_mean_df])
```

```python
pd.concat([yolo8n_mean_df, yolo11s_mean_df])
```
